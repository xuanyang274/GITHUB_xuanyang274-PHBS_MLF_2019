```python
from scipy.io import arff
import pandas as pd
```

```python
data = arff.loadarff('4year.arff')
df = pd.DataFrame(data[0])
```

```python
df['bankruptcy'] = (df['class'] == b'1')
```

```python
df.head()
```

```python
df.describe()
```

```python
sum(df.bankruptcy == True)
```

```python
df2 = df.iloc[:, [0,1,6,9,64,65]]
```

```python
df2.head()
```

```python
df2_col = ['X1', 'X2', 'X7', 'X10', 'class', 'bankruptcy']
df2.columns = df2_col
df2.head()
```

```python
df2.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imr = SimpleImputer(strategy = 'mean')
imr = imr.fit(df2.values)
imputed_data = imr.transform(df2.values)
```

```python
df2.iloc[:, [0,1,2,3]].mean()    #find the mean of the four features among all companies
```

```python
df2.iloc[:, [0,1,2,3]].std()      #find the std of the four features among all companies
```
